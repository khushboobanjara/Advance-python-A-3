```python
# print the first element
import numpy as np
arr = np.array([10,20,30,40,50])
arr[0]
```

```python
# print the last element
arr = np.array([10,20,30,40,50])
arr[-1]
```

```python
# print element from 1 to 3 index
arr = np.array([10,20,30,40,50])
arr[1:4]
```

```python
# reverse the array
arr[::-1]
```

```python
# print the second row
matrix = np.array([[1,2,3],
                [4,5,6]])
matrix[0,]
```

```python
# create matrix of 2X3 print shape
matrix = np.array([[1,2,3],
                [4,5,6]])
matrix.shape
```

```python
# print the first column
matrix[0:,:1]
```

```python
#  Extract elements starting from index 1 with step 2
arr[1::2]
```

```python
#  Extract last 2 elements
arr = np.array([10,20,30,40,50])
arr[-2:]
```

```python
# Extract all elements except last one
arr[:-1]
```

```python
# Reverse using negative slicing
arr[::-1]
```

```python
# Extract values greater than 25
arr[arr>25]
```

```python
# Replace values greater than 25 with 0

arr[arr>25] = 0
print(arr)
```

```python
# Broadcasting
arr = np.array([[10,20,30],
            [40,50,60]])
arr1 = np.array([1,2,3])
arr+arr1
```

```python
arr = np.array([[10,20,30],
            [40,50,60]])
arr1 = np.array([[1],[2]])
arr+arr1
```

```python
# find total sum
arr = np.array([[1,2,3],
[4,5,6]])
arr.sum()
```

```python
# Find column-wise sum
arr.sum(axis=0)
```

```python
# find row wise sum
arr.sum(axis=1)
```

```python
# Mean of each row
arr.mean(axis=1)
```

```python
arr.max(axis=0)
```
